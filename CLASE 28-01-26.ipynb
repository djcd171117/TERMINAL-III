```python
def factorial(n):
  if n == 0 or n == 1: #Caso Base
    return 1
  else:
      return n* factorial(n-1)
```

```python
factorial (5)
```

#SERIE DE FIBONACCI: 0 1 1 2 3 5 8 ...

```python
def fibonacci(n):
  if n == 0:
        return n
  elif n==1:
        return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(8)
```
